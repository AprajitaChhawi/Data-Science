```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def sigmoid(h):
    return 1/(1+2.71**(-h))
```

```python
def sigmoid_der(h):
    return h*(1-h)
```

```python
X = [[0, 0], [0, 1], [1, 0], [1, 1]]
Y = [[0], [0], [0], [1]]
W=np.asarray([0.5,-0.2])
W=W.reshape(2,1)
B =[-0.1]
err=[]
for i in range(0,500):
    a1 = np.asarray(X)
    z2 = np.dot( a1, W) + B
    a2 = sigmoid(z2)
    a2=np.asarray(a2)
    error = (Y - a2)
    delta_output = error * sigmoid_der(a2)
    update = np.dot(a1.T, delta_output)
    W = W + update
    B = B + sum(delta_output)
    err.append(sum(error)/4)
```

```python
plt.plot(range(0,500),err, 'b')
plt.show()
```
