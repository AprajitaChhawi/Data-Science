```python
from math import * #FUNCTION FOR EUCLIDEAN-DISTANCE
def euclidean_distance(x,y):
    return sqrt(sum(pow(a-b,2) for a,b in zip(x,y)))
```

```python
l1=[]
l2=[]
count_eucd=0
for index,row in x2.iterrows():
    eucd_train1_test1=0
    eucd_train2_test1=0
    i=list(row)
    j=i[:2]
    eucd_train1_test1=euclidean_distance(j,center1)
    eucd_train2_test1=euclidean_distance(j,center2)
    min1=min(eucd_train1_test1,eucd_train2_test1)
    if(min1==eucd_train1_test1):
        l1.append('w1')
    elif(min1==eucd_train2_test1):
        l1.append('w2')
x2['Predicted']=l1
for index,row in x2.iterrows():
    i=list(row)
    if(i[2]==i[3]):
        l2.append(0)
    else:
        l2.append(1)
x2['Error']=l2
```

```python
for index,row in x2.iterrows():
    i=list(row)
    if(i[4]==1):
        count_eucd=count_eucd+1
print("The Misclassification rate of euclidean distance classifier is : ",count_eucd/5000)
```
